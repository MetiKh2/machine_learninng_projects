```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
```

```python
data=pd.read_csv('insurance.csv')
data
```

```python
def classify(col,values,df):
  for i,val in enumerate(values):
     df.loc[df[col] == val,[col]] =i

```

```python
classify(col='sex',values=['male','female'],df=data)
classify(col='smoker',values=['yes','no'],df=data)
classify(col='region',values=['southeast','southwest','northeast','northwest'],df=data)
```

```python
data
```

```python
x=data.drop('charges',axis=1)
y=data['charges']
scaler = StandardScaler()
scaler.fit(x)
x=scaler.transform(x)
y
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=2)
 
print(x_test.shape)
print(y_test.shape)
print(x_train.shape)
print(y_train.shape)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
ln_model=LinearRegression()
ln_model.fit(x_train,y_train)
pred=ln_model.predict(x_test)
r2_score(y_test,pred)
```
