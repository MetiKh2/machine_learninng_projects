```python
import numpy as np
import pandas as pd
```

```python
data=pd.read_csv("car.data",names=['buying','maint','doors','persons','lug_boot','safety','class'])
```

```python
data
```

```python
def classify(col,values):
  for i,val in enumerate(values):
     data.loc[data[col] == val,[col]] =i
  # print(data)
```

```python
# data.loc[data['buying'] == 'vhigh',['buying']] =1
classify(col='buying',values=['low','med', 'high','vhigh'])
classify(col='maint',values=['low','med', 'high','vhigh'])
classify(col='lug_boot',values=['small','med', 'big'])
classify(col='safety',values=['low','med', 'high'])
classify(col='class',values=['unacc','acc', 'good','vgood'])

```

```python
data
```

```python
data.loc[data['persons'] == 'more',['persons']] =6
```

```python
data
```

```python
data.loc[data['doors'] == '5more',['doors']] =5
```

```python
data
```

```python
from sklearn.model_selection import train_test_split 
x=data.drop('class',axis=1)
y=data["class"]
x=x.astype('int')
y=y.astype('int')
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=41) 
print("Shape of the X_train", x_train.shape) 
print("Shape of the y_train", y_train.shape) 
print("Shape of the X_test", x_test.shape) 
print("Shape of the y_test", y_test.shape) 
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn_model=KNeighborsClassifier(n_neighbors=5)
knn_model.fit(x_train,y_train)
```

```python
pred=knn_model.predict(x_test)
```

```python
pred
```

```python
y_test.to_list()
```

```python
from sklearn.metrics import classification_report,accuracy_score
print(classification_report(y_test,pred))
```
