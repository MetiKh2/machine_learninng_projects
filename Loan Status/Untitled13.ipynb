```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('data.csv')
```

```python
data
```

```python
def classify(col,values,data_set):
  for i,val in enumerate(values):
     data_set.loc[data_set[col] == val,[col]] =i
```

```python
classify(col='Gender',values=['Male','Female'],data_set=data)
classify(col='Married',values=['No','Yes'],data_set=data)
classify(col='Education',values=['Graduate','Not Graduate'],data_set=data)
classify(col='Self_Employed',values=['No','Yes'],data_set=data)
classify(col='Property_Area',values=['Rural','Semiurban','Urban'],data_set=data)
classify(col='Loan_Status',values=['N','Y'],data_set=data)
data.loc[data['Dependents'] == '3+',['Dependents']] =3


```

```python
data
```

```python
data.isnull().sum()
```

```python
data=data.dropna()
```

```python
# data=data.astype('int')
```

```python
x=data.drop('Loan_Status',axis=1)
x=x.drop('Loan_ID',axis=1)
y=data['Loan_Status']
y=y.astype('int')
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=41)
print("Shape of the X_train", x_train.shape) 
print("Shape of the y_train", y_train.shape) 
print("Shape of the X_test", x_test.shape) 
print("Shape of the y_test", y_test.shape) 
```

```python
lg_model=LogisticRegression()
lg_model.fit(x_train,y_train)
lg_pred=lg_model.predict(x_test)
accuracy_score(y_test,lg_pred)
```

```python
from sklearn.naive_bayes import GaussianNB

nb_model=GaussianNB()
nb_model.fit(x_train,y_train)
nb_pred=nb_model.predict(x_test)
accuracy_score(y_test,nb_pred)
```
